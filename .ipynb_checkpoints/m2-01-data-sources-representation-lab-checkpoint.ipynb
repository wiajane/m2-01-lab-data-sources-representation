Step 3: Write a validation function called validate_registrations(entries) that takes Dataset A and returns two lists: valid_entries and invalid_entries. A valid entry must have a non-empty username, an email containing @, and an age that can be converted to a positive integer. Test your function on Dataset A and print the counts.

```python
def validate_registrations(entries):
    valid_entries = []
    invalid_entries = []

    for entry in entries:
        try:
            username_valid = bool(entry["username"])
            email_valid = "@" in entry["email"]
            age_valid = int(entry["age"])> 0

            if username_valid and email_valid and age_valid:
                valid_entries.append(entry)
            else:
                invalid_entries.append(entry)

        except:
            invalid_entries.append(entry)

    return valid_entries, invalid_entries


valid, invalid = validate_registrations(dataset_a)
print("Valid entries:", len(valid))
print("Invalid entries:", len(invalid))
```

## TASK 2

Step 1: Create a pandas DataFrame called rides with 500 rows and the following columns:

```python
import pandas as pd
import numpy as np

np.random.seed(64)

rides = pd.DataFrame({
    "ride_id": range(1, 501),
    "driver_id": np.random.randint(1000, 1051, 500),
    "distance_km": np.random.uniform(1.0, 30.0, 500),
    "fare_usd": np.random.uniform(5.0, 75.0, 500),
    "ride_type": np.random.choice(["standard", "premium", "pool"], 500),
    "city": np.random.choice(["Berlin", "Seoul", "Nairobi", "Toronto", "Lima"], 500),
    "timestamp": pd.date_range("2025-01-01", periods=500, freq="min")
})
```

Step 2: Save the DataFrame in three formats:

```python
rides.to_csv("rides.csv", index=False)
rides.to_json("rides.json", orient="records", lines=True)
rides.to_parquet("rides.parquet", index=False)
```

Step 3: Compare the file sizes of all three formats. Print the size of each file in KB and compute how many times smaller Parquet is compared to CSV.

```python
import os

csv_size = os.path.getsize("rides.csv") / 1024
json_size = os.path.getsize("rides.json") / 1024
parquet_size = os.path.getsize("rides.parquet") / 1024

print(f"CSV size: {csv_size:.2f} KB")
print(f"JSON size: {json_size:.2f} KB")
print(f"Parquet size: {parquet_size:.2f} KB")

print("Parquet is", round(csv_size / parquet_size, 2), "times smaller than CSV")
```

I used OS modue because os module lets us interact with the operating system besides checking file metadata like size (in bytes).

Step 4: Read each file back into a new DataFrame and verify that the data is identical by comparing shapes and a sample of values. Note any differences you observe (for example, type changes after reading from CSV).

```python
df_csv = pd.read_csv("rides.csv")
df_json = pd.read_json("rides.json", orient="records", lines=True)
df_parquet = pd.read_parquet("rides.parquet")

print(df_csv.shape == rides.shape)
print(df_json.shape == rides.shape)
print(df_parquet.shape == rides.shape)

print(df_csv.head().equals(rides.head()))
print(df_parquet.head().equals(rides.head()))
```

.equals() is strict. Even small dtype differences will return False. When reading from CSV, timestamp column will become: object instead of its oirignal type (datetime).

Step 5: Write a markdown cell summarizing: Which format preserved types most faithfully? Which was smallest? When would you choose each format?

1. Parquet preserved data types most faithfully. It maintained numeric types and the datetime64 type for the timestamp column without requiring additional arguments during reading

2. Parquet was the smallest file. Because it is a compressed binary, columnar format, it stores data more efficiently than CSV or JSON.

## TASK 3

Step 1: Create a large DataFrame with 200,000 rows and 20 numeric columns (random float values). Name the columns feature_0 through feature_19.

```python
df = pd.DataFrame(
    np.random.rand(200000, 20),
    columns=[f"feature_{i}" for i in range(20)]
)
```

Step 2: Measure and compare the time for these two operations:

Operation A — Column access: Compute the mean of feature_0 using the pandas column accessor (df["feature_0"].mean()).

Operation B — Row iteration: Loop through the first 10,000 rows using df.iloc[i] and sum the values of feature_0 manually.

Use %%timeit or %time to measure each. Record the results.

```python
%timeit df["feature_0"].mean()

def row_loop(df):
    total = 0
    for i in range(10000):
        total += df.iloc[i]["feature_0"]
    return total

%timeit row_loop(df)
```

Step 3: Now convert the DataFrame to a NumPy array and repeat both operations:

Operation C — NumPy column access: Compute the mean of column 0 (arr[:, 0].mean()).

Operation D — NumPy row access: Access the first 10,000 rows sequentially (arr[i] in a loop) and sum column 0.

Record the results.

```python
arr = df.to_numpy()

%timeit arr[:, 0].mean()

def numpy_row_loop(arr):
    total = 0
    for i in range(10000):
        total += arr[i][0]
    return total

%timeit numpy_row_loop(arr)
```

Step 4: Write a markdown cell that answers:

Why is column access in pandas so much faster than row iteration?
Why does the gap narrow (or change) when using NumPy?
What does this tell you about how pandas and NumPy lay out data in memory?

1. Pandas is optimized for column-wise operations, so accessing a full column uses fast, vectorized computations. Row iteration is slower because each row access creates a pandas Series object and adds Python-level overhead.
2. NumPy arrays are stored in contiguous memory, so both row and column access are more efficient than in pandas. The performance gap changes because NumPy avoids the extra object overhead that pandas introduces.
3. It shows that both libraries are optimized for column-based operations, especially pandas. NumPy stores data in compact, homogeneous arrays, while pandas builds on NumPy but adds extra structure that makes row access slower.

## TASK 4

Step 1: Create a DataFrame with 100,000 rows and 5 numeric columns of random float values (use np.random.seed(42) for reproducibility). Add one string column called category with values randomly chosen from ["alpha", "beta", "gamma"].

```python
np.random.seed(42)

df_large = pd.DataFrame(
    np.random.rand(100000, 5),
    columns=[f"num_{i}" for i in range(5)]
)

df_large["category"] = np.random.choice(["alpha", "beta", "gamma"], 100000)
```

Step 2: Save as CSV and Parquet. Compare file sizes.

```python
df_large.to_csv("large.csv", index=False)
df_large.to_parquet("large.parquet", index=False)
```

Step 3: Read the CSV back and check the dtypes. Then read the Parquet back and check the dtypes. Write a markdown cell explaining any differences you observe in how types were preserved.

```python
df_csv = pd.read_csv("large.csv")
print("CSV dtypes:")
print(df_csv.dtypes)

df_parquet = pd.read_parquet("large.parquet")
print("Parquet dtypes:")
print(df_parquet.dtypes)
```

After reading the files back, the numeric columns were preserved correctly in both formats. However, CSV stores data as plain text, so pandas must infer the data types when reading it.
Parquet preserved the schema more faithfully because it stores type metadata in the file. This makes Parquet more reliable for maintaining exact data types, especially for complex types like datetimes or categoricals.

Step 4: Perform a timing comparison: measure how long it takes to read the full dataset from CSV versus from Parquet. Record both times.

```python
%timeit pd.read_csv("large.csv")
%timeit pd.read_parquet("large.parquet")
```

Step 5: Parquet supports reading only specific columns without loading the entire file. Read only the category column from the Parquet file and time it. Compare to reading the full file. Write a markdown cell explaining why this selective read is possible with Parquet but not with CSV.

```python
%timeit pd.read_parquet("large.parquet")
%timeit pd.read_parquet("large.parquet", columns=["category"])
```

Parquet is a columnar storage format, meaning data is stored column-by-column on disk. It can read only the requested column without loading the rest of the dataset.
CSV is a row-based text format, where each row contains all columns together. To read one column, the entire file must still be scanned and parsed, which makes selective column reading inefficient or impossible.
